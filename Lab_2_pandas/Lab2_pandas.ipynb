### 3. Список наименований средств, соответствующих двум любым документам с требованиями по сертификации

```python
def find_2_docs(str):
    """Функция нахождения количества запятых в строке"""
    
    p = re.compile(',')
    coma_len = len(p.findall(str))
    
    return coma_len
```

```python
coma_len = df['Наименования документов, требованиям которых соответствует средство'].apply(find_2_docs)
```

В новый столбец добавляем показатель количества запятых в строке

```python
df['coma_len'] = coma_len
```

Отбираем те значения, где количество запятых равно 1, то есть количество документов, требованиям которых соответствует средство равно 2

### 4. Наименования средств, которые имеют просроченные сертификаты

```python
# Удаление пустых значений
df_date_term = df.dropna(subset = 'Срок действия сертификата')
str_date_term = df_date_term[df_date_term['Срок действия сертификата']=='#Н/Д'].index
df_date_term.drop(str_date_term, axis=0, inplace=True)
```

```python
# Перевод в формат даты "Срок действия сертификата"
df_date_term['Срок действия сертификата'] \
     = pd.to_datetime(df_date_term['Срок действия сертификата'])
```

```python
def return_overdue():
    """Средства, которые имеют просроченные сертификаты"""
    
    date_today = pd.to_datetime(datetime.date.today())
    df_overdue = df_date_term[df_date_term['Срок действия сертификата'] < date_today]
    
    return df_overdue
```

```python
# Наименования средств, которые имеют просроченные сертификаты
funds_overdue = return_overdue()['Наименование средства (шифр)']
funds_overdue
```

```python
funds_overdue.shape
```

### 5. Наименования средств, которые имеют просроченные сертификаты, но при этом обязаны оказывать поддержку

```python
# Удаление пустых значений
df_date = return_overdue().dropna(subset = 'Информация об окончании срока технической поддержки, полученная от заявителя')
str_date = df_date[df_date['Информация об окончании срока технической поддержки, полученная от заявителя']=='бессрочно']
df_date.drop(str_date.index, axis=0, inplace=True)
```

```python
str_date
```

```python
# Перевод в формат даты
df_date['Информация об окончании срока технической поддержки, полученная от заявителя'] \
     = pd.to_datetime(df_date['Информация об окончании срока технической поддержки, полученная от заявителя'])
```

```python
def return_overdue_and_support():
    """Средства, которые имеют просроченные сертификаты, но должны оказывать поддержку"""
    date_today = pd.to_datetime(datetime.date.today())
    df_overdue = df_date[df_date['Информация об окончании срока технической поддержки, полученная от заявителя']>date_today]
    
    return df_overdue
```

```python
funds_overdue_and_support = return_overdue_and_support()['Наименование средства (шифр)']
funds_overdue_and_support
```

```python
def find_private_organiz(str):
    
    private_organiz_exist = str.find("ООО ") * str.find("ЗАО ") * str.find("ПАО ") * str.find("АО ")
    
    return private_organiz_exist
```

```python
private_organiz_exist = df['Испытательная лаборатория'].apply(find_private_organiz)
```

```python
df['private_organiz_exist'] = private_organiz_exist
```
